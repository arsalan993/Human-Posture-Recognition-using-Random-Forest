```python
# Load the library with the iris dataset
from sklearn.datasets import load_iris
from sklearn.metrics import f1_score
# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier

# Load pandas
import pandas as pd

# Load numpy
import numpy as np

# Set random seed
np.random.seed(0)
```

```python
df = pd.read_csv('smartphoneatpocket.csv',header=None)
```

```python
df.head()
```

```python
df = df.drop(df.columns[[0]], axis=1)
df.head()
```

```python
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75

# View the top 5 rows
df.head()
```

```python
# Create two new dataframes, one with the training rows, one with the test rows
train, test = df[df['is_train']==True], df[df['is_train']==False]
# Show the number of observations for the test and training dataframes
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:',len(test))
```

```python
df.head()
```

```python
# Create a list of the feature column's names
features = df.columns[:12]

# View features
features
```

```python
y = train[13]
```

```python
clf = RandomForestClassifier(n_jobs=100,n_estimators = 100, random_state=0, min_samples_split=6, min_samples_leaf=2,oob_score=True,max_features='log2')

# Train the Classifier to take the training features and learn how they relate
# to the training y (the species)
clf.fit(train[features], y)
```

```python
# Apply the Classifier we trained to the test data (which, remember, it has never seen before)
preds =  clf.predict(test[features])
```

```python
# Create confusion matrix
pd.crosstab(test[13], preds, rownames=['Actual'], colnames=['Predicted'])
```

```python
#‘binary’ (default), ‘micro’, ‘macro’, ‘samples’, ‘weighted
print(f1_score(test[13], preds, average='macro') )

print(f1_score(test[13], preds, average='micro')  )

print(f1_score(test[13], preds, average='weighted') ) 
```
